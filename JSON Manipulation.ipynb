```python
import pandas as pd

# read in dataset
df = pd.read_json('https://raw.githubusercontent.com/jasonwnc/ds2002s24/main/2.%20Python/coll%20bb_SortByConf.json')

df = pd.json_normalize(df['teams'])

# view first 100 rows
print(df.head(100))

# total teams:
num_teams = df['region'].nunique()
print(f'Total number of teams: {num_teams}')

# total teams in virginia:
teams_in_virginia = df[df['state'] == 'VA']['region'].nunique()
print(f'Teams in Virginia: {teams_in_virginia}')

# list of duplicate mascots:
mascot_counts = df['name'].value_counts()
duplicate_mascots = mascot_counts[mascot_counts > 1]
print('Duplicate mascots grouped by number:')
print(duplicate_mascots)
```
